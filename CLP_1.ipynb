```python
import pandas as pd

test_data = pd.read_csv('/content/test.csv')
display(test_data)
```

```python
train_data = pd.read_csv('/content/train.csv')
display(train_data)
```

```python
mer_data = pd.concat([train_data, test_data], ignore_index=True)
display(mer_data.shape)
```

```python
mer_data.columns
```

```python
mer_data.columns = mer_data.columns.str.strip()
```

```python
data = mer_data
```

```python
data.info()
```

```python
columns_to_keep = ['Street', 'LandContour', 'Neighborhood', 'HouseStyle', 'RoofStyle', 'Heating', 'Electrical', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'PoolArea', 'SalePrice']
data = mer_data[columns_to_keep]
display(data.head())
```

```python
import numpy as np

print("Null values per column:")
display(data.isnull().sum())

print("\nInfinite values per column:")
display(data.isin([np.inf, -np.inf]).sum())
```

```python
# Fill missing 'Electrical' values with the mode
data['Electrical'].fillna(data['Electrical'].mode()[0], inplace=True)

print("Null values per column after filling Electrical:")
display(data.isnull().sum())
```

```python
categorical_cols = data.select_dtypes(include='object').columns
data = pd.get_dummies(data, columns=categorical_cols)
display(data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data['SalePrice'].dropna(), kde=True, bins=50)
plt.title('Distribution of Sale Prices')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_cols = data.select_dtypes(include=np.number).columns
correlation_matrix = data[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get all columns that start with 'Neighborhood_'
neighborhood_cols = [col for col in data.columns if col.startswith('Neighborhood_')]

# Iterate through each one-hot encoded neighborhood column and create a box plot
for col in neighborhood_cols:
    print(f"Displaying box plot for: {col}")
    plt.figure(figsize=(10, 6))
    # Filter the data to include only rows where the current neighborhood is True
    sns.boxplot(x=col, y='SalePrice', data=data[data[col]].dropna(subset=['SalePrice']))
    plt.title(f'Box Plot of {col} vs. SalePrice')
    plt.xlabel(col)
    plt.ylabel('Sale Price')
    plt.xticks(rotation=45)
    plt.show()
```

```python
X = data.drop('SalePrice', axis=1)
y = data['SalePrice']
```

```python
from sklearn.preprocessing import StandardScaler



numerical_cols = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()

X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

display(X.head())
```

```python
from sklearn.model_selection import train_test_split

# Separate features (X) and target (y) for the original training data
X_train_full = X[y.notna()]
y_train_full = y[y.notna()]
X_test_submission = X[y.isna()] # This is the original test set for submission

# Use train_test_split on the original training data to create training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of X_val:", X_val.shape)
print("Shape of y_val:", y_val.shape)
print("Shape of X_test_submission:", X_test_submission.shape)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model architecture
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='tanh'),
    Dense(32, activation='relu'),
    Dense(1)  # Output layer for regression
])

# Compile the model
#Parameters = (Number of input features × Number of neurons) + Number of neurons
model.compile(optimizer='adam', loss='mse')

model.summary()
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# Plot training & validation loss curves
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Make predictions on the training data
y_train_pred = model.predict(X_train)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
print(f"Root Mean Squared Error (RMSE) on training data: {rmse:.2f}")

# Calculate MAE
mae = mean_absolute_error(y_train, y_train_pred)
print(f"Mean Absolute Error (MAE) on training data: {mae:.2f}")
```
